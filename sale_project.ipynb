```python
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_csv('dataset.csv')

#
def plot_coordinates(df, start_date_str, end_date_str):
    # Assuming the date format in the CSV is '1402-12-09'
    # If the format is different, adjust accordingly
    start_date = start_date_str
    end_date = end_date_str

    names = df['name'].unique()
    dfs = []
    for name in names:
        df_temp = df[df['name'] == name]
        df_temp = df_temp[(df_temp['date'] >= start_date) & (df_temp['date'] <= end_date)]
        dfs.append(df_temp)
    plt.figure(figsize=(6, 4))
    for i, df_temp in enumerate(dfs):
        plt.plot(df_temp['date'], df_temp['sale'], label=names[i])

    plt.xlabel('date')
    plt.ylabel('sale')
    plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.xticks(rotation=90)
    plt.show()

#
df0=df.drop(['count','price','name'], axis='columns')
df_merged = df0.groupby('date' , as_index=False).sum()
def total_sales(df_merged, start_date_str, end_date_str):
    plt.figure(figsize=(15, 10))
    start_date = start_date_str
    end_date = end_date_str
    # Filter the dataframe based on the start and end dates
    filtered_df = df_merged[(df_merged['date'] >= start_date) & (df_merged['date'] <= end_date)]
    
    # Plot the graph
    plt.plot(filtered_df['date'], filtered_df['sale'])
    plt.xlabel('Date')
    plt.ylabel('Sale')
    plt.title('month total sale')
    plt.grid(color='green', linestyle='--', linewidth=0.5)
    plt.xticks(rotation=90)
    plt.show()

#
def weeks_sale():
    start_date = i_1
    end_date = i_2
    plot_coordinates(df, start_date, end_date)

    start_date = i_3
    end_date = i_4
    plot_coordinates(df, start_date, end_date)

    start_date = i_5
    end_date = i_6
    plot_coordinates(df, start_date, end_date)

    start_date = i_7
    end_date = i_8
    plot_coordinates(df, start_date, end_date)

    start_date = i_9
    end_date = i_10
    plot_coordinates(df, start_date, end_date)

#
print('Hello!')
print('Enter "n" for input new data: ')
print('Enter "m" to check monthly sale: ')
print('Enter "w" to check weekly sale: ')
print('Enter "t" to check total sale ')

input_0=str(input('What do you want do do: '))

if input_0== 'n':
    x = list(map(str, input("Enter your new data : ").split()))
    n = x[0]
    c = int(x[1])
    p = int(x[2])
    d = x[3]
    s=p*c
    new_row = pd.Series([n, c, p, d,  s ], index=['name', 'count', 'price','date','sale' ])
    df = pd.concat([df, new_row.to_frame().T], ignore_index=True)
    df.to_csv("dataset.csv",index=False)
    print(n , c, p , d)
if input_0== 'm':
     
         
    start_date=str(input('Enter start date: '))
    end_date=str(input('Enter end date: '))
    names=df['name'].unique()
    df2=df.drop(['count','price'], axis='columns')
    plot_coordinates(df, start_date, end_date)
if input_0== 'w':
    start=str(input('Enter start date: '))
    end=str(input('Enter end time: '))
    df_mergeds = df_merged[df_merged['date'].between(start, end)]
    i_1=str(df_mergeds['date'].unique()[0])
    i_2=str(df_mergeds['date'].unique()[6])
    i_3=str(df_mergeds['date'].unique()[7])
    i_4=str(df_mergeds['date'].unique()[13])
    i_5=str(df_mergeds['date'].unique()[14])
    i_6=str(df_mergeds['date'].unique()[20])
    i_7=str(df_mergeds['date'].unique()[21])
    i_8=str(df_mergeds['date'].unique()[27])
    i_9=str(df_mergeds['date'].unique()[28])
    i_10=str(df_mergeds['date'].unique()[-1])
    weeks_sale()
        
if input_0== 't':
    start_date=str(input('Enter start date: '))
    end_date=str(input('Enter end date: '))
    total_sales(df_merged, start_date, end_date)
        
```
